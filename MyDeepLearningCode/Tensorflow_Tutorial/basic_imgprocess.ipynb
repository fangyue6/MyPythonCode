```python
"""
Basic image load,plot,resize,etc...
author zhenlei Xu(hadxu123@gmail.com)
"""
import numpy as np
import tensorflow as tf
import pandas 
import skimage.io
import skimage.transform
import matplotlib.pyplot as plt
from scipy.misc import imread,imresize
import os

%matplotlib inline
print 'package loaded'
```

```python
#print current folder
cwd = os.getcwd()
print cwd
```

```python
cat = imread(cwd+'/images/cat.jpg')
print type(cat),cat.shape
plt.figure(0)
plt.imshow(cat)
plt.title('origin iamge')
plt.draw()
```

```python
#resize
catsmall = imresize(cat,[100,100,3])
plt.figure(0)
plt.imshow(catsmall)
plt.title('Resized image')
plt.draw()
```

```python
#grayscale
def rgb2gray(rgb):
    return np.dot(rgb[...,:3],[0.299,0.587,0.114])
catsmallgray = rgb2gray(catsmall)
plt.figure(2)
plt.imshow(catsmallgray,cmap=plt.cm.gray)
plt.title('gray')
plt.colorbar()
plt.draw()
```

```python
#convert to Vector
print catsmallgray.shape
print '-----------------'
catrowvec = np.reshape(catsmallgray,(1,-1))
print catrowvec.shape
```

```python
#convet to matrix
catmatrix = np.reshape(catrowvec,(100,100))
print catmatrix.shape
```

```python
#load from folder
cwd = os.getcwd()
path = cwd+'/images'
valid_exts = ['.jpg','.gif','.png','.tga']
print 'images in %s are: \n %s' %(path,os.listdir(path))
imgs =[]
names = []
for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_exts:
        continue
    fullpath = os.path.join(path,f)
    imgs.append(imread(fullpath))
    names.append(os.path.splitext(f)[0]+os.path.splitext(f)[1])
print type(imgs)
print len(imgs)
for curr_img,curr_name,i in zip(imgs,names,range(len(imgs))):
    print i,type(curr_img)
    print curr_name
    print curr_img.shape
```
